```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\srivathsa\Desktop\EV_cars.csv")  # Replace 'your_dataset.csv' with the actual filename



```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Fast_charge', data=df)
plt.title('Frequency of Fast Charging')
plt.xlabel('Fast Charging')
plt.ylabel('Count')
plt.show()
```

```python
# Heatmap to visualize correlations between numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Group the data by Fast_charge and calculate the average efficiency for each category
efficiency_by_charge = df.groupby('Fast_charge')['Efficiency'].mean().reset_index()

# Plotting the average efficiency for each fast charging category
plt.figure(figsize=(10, 6))
sns.lineplot(data=efficiency_by_charge, x='Fast_charge', y='Efficiency', marker='o', color='blue')
plt.title('Average Efficiency by Fast Charging Type')
plt.xlabel('Fast Charging Type')
plt.ylabel('Average Efficiency')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
